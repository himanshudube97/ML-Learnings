What is Correlation?
Correlation measures the relationship between two variables.
Positive Correlation (closer to +1): When one variable increases, the other also increases.
Negative Correlation (closer to -1): When one variable increases, the other decreases.
No Correlation (around 0): No clear relationship between the two variables.
For example:

Height and weight may have a positive correlation (taller people often weigh more).
Exercise and body fat may have a negative correlation (more exercise, less fat).
Correlation Matrix
A correlation matrix is a table showing the correlation between all pairs of variables in your dataset.
Itâ€™s a grid where:
Rows and columns represent variables.
Each cell contains the correlation value between two variables.
Values range from -1 (strong negative) to +1 (strong positive).
Example:

Height	Weight	Age
Height	1.0	0.8	0.1
Weight	0.8	1.0	0.2
Age	0.1	0.2	1.0
Diagonal cells (1.0): These are always 1.0 because a variable is perfectly correlated with itself.
Off-diagonal cells: Show correlations between different variables

```python
# Matrix form for correlation data
data_1 = tips.corr(numeric_only=True)
data_1
```

```python
sns.heatmap(data_1,annot=True)
plt.show()
```

```python
## rearranging flights data
fp = flights.pivot_table(values='passengers',index='month',columns='year', observed=False)
fp
```

```python
sns.heatmap(fp, cmap="coolwarm", linewidths=1)
plt.show()
```
