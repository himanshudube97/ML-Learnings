```python
import numpy as np
import pandas as pd
```

```python
data = {
    'Company':["Goog", "Goog", "Msft", "Msft", "Fb", "Fb"],
    "Person": ["Sam", "Charlie", "Amy", "Vanessa", "Carl", "Sarah"],
    "Sales":[200, 120, 340, 124, 243, 350]
}
```

```python
df = pd.DataFrame(data)
df
```

```python
byComp = df.groupby("Company")
byComp
```

```python
byComp.mean("Sales")  ## in the video it said that pandas automatically removes item while aggregation if its non-numeric., but i dont think it s happening here.
```

```python
byComp.sum()  ## see this without specifying the column it did a concatenation of the person column
```

```python
byComp.sum("Sales")
```

```python
## This one is important as it gives information about everything
df.groupby("Company").describe()
```

```python
##change the above format using transpose
df.groupby("Company").describe().transpose()
```
