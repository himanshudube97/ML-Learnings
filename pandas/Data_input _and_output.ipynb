```python
### Reading SQL
## Pandas is not a good way to read sql, because of the different engines that these databases have, postgres, mysql etc
```

```python
## so actually work with specific sql database we need to add the specific driver. (HIGH LEVEL)
## right now we will create a small dummy sql engine
```

```python
from sqlalchemy import create_engine
```

```python
engine = create_engine('sqlite:///:memory:') ## this created a in memory sqlite engine database 
```

```python
df = pd.read_csv("example")
df
```

```python
df.to_sql("my_table", engine)
```

```python
sqldf = pd.read_sql("my_table", con=engine)
```
